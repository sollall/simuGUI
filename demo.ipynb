```python
import time
from ipywidgets import interact, IntSlider, RadioButtons

import threading
from IPython.display import display
import ipywidgets as widgets

widgets = [IntSlider(min=0, max=400),IntSlider(min=0, max=400),widgets.Select(options=['Apple', 'Banana', 'Cherry'])]

def f(ws):
    while True:
        time.sleep(1)
        print([w.value for w in ws])

thread = threading.Thread(target=f, args=(widgets,))

display(*widgets)
thread.start()
```
